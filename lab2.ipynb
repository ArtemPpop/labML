```python
import math
class ComplexNumber:
    def __init__(self, real=0.0, imag=0.0):
        self.real = real
        self.imag = imag
    def multiply_by_scalar(self, scalar):
        return ComplexNumber(self.real * scalar, self.imag * scalar)
    def get_argument(self):
        return math.degrees(math.atan2(self.imag, self.real))
    def format_info(self):
        sign = '+' if self.imag >= 0 else ''
        return f"{self.real}{sign}{self.imag}i ({self.get_argument():.2f}градусов)"
c1 = ComplexNumber(3, 4)
print("Исходное:", c1.format_info())
scalar = float(input("Введите скаляр: "))
c2 = c1.multiply_by_scalar(scalar)
print("Результат:", c2.format_info())
real = float(input("Введите первое: "))
imag = float(input("Введите второе: "))
c3 = ComplexNumber(real, imag)
print("Итог:", c3.format_info())
```

```python
import math

class ComplexNumber:
    def __init__(self, real=0.0, imag=0.0):
        self.real = real   # *действительная часть*
        self.imag = imag   # *мнимая часть*

    def modulus(self):
        return math.sqrt(self.real**2 + self.imag**2)

    def argument(self):
        return math.degrees(math.atan2(self.imag, self.real))

    def reciprocal(self):
        denom = self.real**2 + self.imag**2
        if denom == 0: raise ZeroDivisionError("Деление на ноль")
        return ComplexNumber(self.real/denom, -self.imag/denom)
class RLBranch(ComplexNumber):
    def __init__(self, R, L, f=50.0):
        self.R = R          # своё поле
        self.L = L          # своё поле
        self.f = f          # своё поле
        self.omega = 2 * math.pi * f
        super().__init__(R, self.omega * L)  # инициализация родителя: Z = R + jωL

    def process_data(self):
        Y = self.reciprocal()  # Y = 1/Z
        return {
            'Y_modulus': Y.modulus(),
            'Y_argument': Y.argument(),
            'Y_real': Y.real,
            'Y_imag': Y.imag
        }

branch = RLBranch(R=100, L=0.1, f=50)
result = branch.process_data()

print(f"★ Модуль проводимости: |Y| = {result['Y_modulus']:.6f} См")
print(f"★ Аргумент проводимости: arg(Y) = {result['Y_argument']:.2f}°")
```

```python
class Diamond:
    """Класс 1-го уровня - Алмаз"""
    def __init__(self, name="", weight=0.0, cut_quality=0.0):
        self.name = name              # название
        self.weight = weight          # вес в каратах
        self.cut_quality = cut_quality # качество огранки в баллах

    def quality_Q(self):
        """Функция качества Q = 0.4·вес + 0.6·качество огранки"""
        return 0.4 * self.weight + 0.6 * self.cut_quality

    def display_info(self):
        """Вывод информации об объекте"""
        print(f"\n{'='*40}")
        print(f"АЛМАЗ: {self.name}")
        print(f"{'='*40}")
        print(f"Вес: {self.weight} карат")
        print(f"Качество огранки: {self.cut_quality} баллов")
```

```python
class ColoredDiamond(Diamond):
    """Класс 2-го уровня - Цветной алмаз (наследник Diamond)"""
    def __init__(self, name="", weight=0.0, cut_quality=0.0, color=""):
        # Вызов конструктора родителя
        super().__init__(name, weight, cut_quality)
        self.color = color  # дополнительное поле P: цвет

    def quality_Qp(self):
        """Перекрытая функция качества Qp с учётом цвета"""
        Q = self.quality_Q()  # получаем качество родителя

        if self.color.lower() == "голубой":
            Qp = Q + 1
        elif self.color.lower() == "желтый":
            Qp = Q - 0.5
        else:
            Qp = Q

        return Qp

    def display_info(self):
        """Перекрытый вывод информации с учётом цвета"""
        super().display_info()  # вывод информации родителя
        print(f"Цвет: {self.color}")
        print(f"Качество с учётом цвета (Qp): {self.quality_Qp():.2f}")
        print(f"{'='*40}\n")
```

```python
print("ПРОГРАММА ДЛЯ РАБОТЫ С АЛМАЗАМИ")
print("="*40)

# Создание объектов класса 1-го уровня (родитель)
print("\n>>> СОЗДАНИЕ АЛМАЗА (класс 1-го уровня):")
name1 = input("Введите название алмаза: ")
weight1 = float(input("Введите вес (в каратах): "))
cut1 = float(input("Введите качество огранки (в баллах): "))

diamond1 = Diamond(name1, weight1, cut1)

# Создание объектов класса 2-го уровня (потомок)
print("\n>>> СОЗДАНИЕ ЦВЕТНОГО АЛМАЗА (класс 2-го уровня):")
name2 = input("Введите название алмаза: ")
weight2 = float(input("Введите вес (в каратах): "))
cut2 = float(input("Введите качество огранки (в баллах): "))
color = input("Введите цвет (белый/голубой/желтый/другой): ")

diamond2 = ColoredDiamond(name2, weight2, cut2, color)

# Вывод информации
print("\n" + "="*40)
print("РЕЗУЛЬТАТЫ:")
print("="*40)

diamond1.display_info()
diamond2.display_info()

# Пример с разными цветами
print("\n>>> ПРИМЕРЫ РАЗНЫХ ЦВЕТОВ:")
examples = [
    ColoredDiamond("Белый алмаз", 2.0, 8.0, "белый"),
    ColoredDiamond("Голубой алмаз", 2.0, 8.0, "голубой"),
    ColoredDiamond("Желтый алмаз", 2.0, 8.0, "желтый")
]
for ex in examples:
    ex.display_info()
```
