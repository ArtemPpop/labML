```python
import math
class ComplexNumber:
    def __init__(self, real=0.0, imag=0.0):
        self.real = real
        self.imag = imag
    def multiply_by_scalar(self, scalar):
        return ComplexNumber(self.real * scalar, self.imag * scalar)
    def get_argument(self):
        return math.degrees(math.atan2(self.imag, self.real))
    def format_info(self):
        sign = '+' if self.imag >= 0 else ''
        return f"{self.real}{sign}{self.imag}i ({self.get_argument():.2f}градусов)"
c1 = ComplexNumber(3, 4)
print("Исходное:", c1.format_info())
scalar = float(input("Введите скаляр: "))
c2 = c1.multiply_by_scalar(scalar)
print("Результат:", c2.format_info())
real = float(input("Введите первое: "))
imag = float(input("Введите второе: "))
c3 = ComplexNumber(real, imag)
print("Итог:", c3.format_info())
```

```python
import math

class ComplexNumber:
    def __init__(self, real=0.0, imag=0.0):
        self.real = real
        self.imag = imag
    def modulus(self):
        return math.sqrt(self.real**2 + self.imag**2)
    def argument(self):
        return math.degrees(math.atan2(self.imag, self.real))
    def reciprocal(self):
        denom = self.real**2 + self.imag**2
        if denom == 0: raise ZeroDivisionError("Деление на ноль")
        return ComplexNumber(self.real/denom, -self.imag/denom)
class RLBranch(ComplexNumber):
    def __init__(self, R, L, f=50.0):
        self.R = R
        self.L = L
        self.f = f
        self.omega = 2 * math.pi * f
        super().__init__(R, self.omega * L)
    def process_data(self):
        Y = self.reciprocal()
        return {
            'Y_modulus': Y.modulus(),
            'Y_argument': Y.argument(),
            'Y_real': Y.real,
            'Y_imag': Y.imag
        }
branch = RLBranch(R=100, L=0.1, f=50)
result = branch.process_data()
print(f"Модуль проводимости: Y = {result['Y_modulus']:.6f} См")
print(f"Аргумент проводимости: argY = {result['Y_argument']:.2f}°")
```

```python
class Diamond:
    def __init__(self, name="", weight=0.0, cut_quality=0.0):
        self.name = name
        self.weight = weight
        self.cut_quality = cut_quality
    def quality_Q(self):
        return 0.4 * self.weight + 0.6 * self.cut_quality
    def display_info(self):
        print(f"алмаз: {self.name}")
        print(f"Вес: {self.weight} карат")
        print(f"Качество огранки: {self.cut_quality} баллов")
```

```python
class ColoredDiamond(Diamond):
    def __init__(self, name="", weight=0.0, cut_quality=0.0, color=""):
        super().__init__(name, weight, cut_quality)
        self.color = color
    def quality_Qp(self):
        Q = self.quality_Q()
        if self.color.lower() == "голубой":
            Qp = Q + 1
        elif self.color.lower() == "желтый":
            Qp = Q - 0.5
        else:
            Qp = Q
        return Qp
    def display_info(self):
        super().display_info()
        print(f"Цвет: {self.color}")
        print(f"Качество (Qp): {self.quality_Qp():.2f}")
```

```python
print("\nсоздание алмащза :")
name1 = input("Введите название алмаза: ")
weight1 = float(input("вес : "))
cut1 = float(input("качество огранки: "))
diamond1 = Diamond(name1, weight1, cut1)
print("\n цветной алмаз:")
name2 = input("Введите название алмаза: ")
weight2 = float(input("вес : "))
cut2 = float(input("качество огранки: "))
color = input("Введите цвет: ")
diamond2 = ColoredDiamond(name2, weight2, cut2, color)
print("резувльтат:")
diamond1.display_info()
diamond2.display_info()
```
