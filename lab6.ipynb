```python
def find_firm(name):
    for f in firms:
        if f["name"].lower() == name.lower():
            years = datetime.now().year - datetime.strptime(f["date"], "%Y-%m-%d").year
            return f"{f['phone']}\n {f['address']}\nСотрудничество: {years} лет"
    return "Фирма не найдена"

# Пример:
print(find_firm("ООО Вектор"))
```

```python
import matplotlib.pyplot as plt

names = [f["name"] for f in firms]
years = [datetime.now().year - datetime.strptime(f["date"], "%Y-%m-%d").year for f in firms]

plt.figure(figsize=(10, 5))
plt.barh(names, years, color='skyblue')
plt.xlabel('Лет сотрудничества')
plt.title('Длительность сотрудничества с фирмами')
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```
