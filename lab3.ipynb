```python
import random

with open('f.txt', 'w') as f:
    for i in range(20):
        f.write(str(random.randint(1, 50)) + '\n')

print("Файл f.txt создан")
```

```python
seen = set()
unique = []

with open('f.txt', 'r') as f:
    for line in f:
        num = int(line.strip())
        if num not in seen:
            seen.add(num)
            unique.append(num)

with open('g.txt', 'w') as g:
    for num in unique:
        g.write(str(num) + '\n')

print("Файл g.txt создан")
```

```python
print("Содержимое файла g.txt:")
with open('g.txt', 'r') as g:
    print(g.read())
```
